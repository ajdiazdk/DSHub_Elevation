```python
# Function to merge raster using SAGA tool
def merge_rasters(input_rasters, output_raster):
     
    # Command to run SAGA's "Mosaic Rasters" tool
    cmd = f"saga_cmd grid_tools 3 -FILE_LIST:{input_rasters} -TYPE:9 -OVERLAP:5 -BLEND_DIST:300 -RESAMPLING:0 -TARGET_DEFINITION:0 -MATCH:3 -BLEND_BND:0 -TARGET_OUT_GRID:\"{output_raster}\""
    
    # Run the command
    process = subprocess.run(cmd, shell=True, capture_output=True, text=True) 
    
    #Print the output of the command
    print(process.stdout)
    
    if process.stderr:
        print(f"Errors: {process.stderr}")
    else:
        print(f"Rasters merged successfully")
```

```python

# Paths to the input rasters to be merged
input_rasters = ".\\list.txt"

```

```python
# Path to the output raster
output_raster = "output_merged.tif"
```

```python
# Merge the rasters
merge_rasters(input_rasters, output_raster)
```
